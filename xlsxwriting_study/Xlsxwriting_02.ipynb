```python
import ole

with ole.open('입사지원서.hwp') as f:
    data = f.get_stream('PrvText').read().decode('utf-16le')

splitData = data.split('><')
splitData   #데이터 확인
```

```python
wantedData = ['성          명', '생  년  월  일', '연    락    처']
wantedDataIndex = []

for i in splitData:
    for j in wantedData:
        if i == j:
            wantedDataIndex.append(splitData.index(i)+1)   # 원하는 데이터는 +1
wantedDataIndex
```

```python
print(splitData[wantedDataIndex[0]],
splitData[wantedDataIndex[1]],
splitData[wantedDataIndex[2]])

for i in wantedDataIndex:
    print(splitData[i])
```

```python
from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser

output_string = StringIO()
with open('이력서.pdf', 'rb') as in_file:
    parser = PDFParser(in_file)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    for page in PDFPage.create_pages(doc):
        interpreter.process_page(page)

output_string.getvalue()
```

```python
splitdata2 = output_string.getvalue().split('\n\n')
# splitdata2.rstrip('\x00') > 공백제거방법 강구, replace..?
splitdata2
```

```python
infodata2 = ['성\x00 \x00 \x00 \x00 \x00 \x00 \x00 \x00 \x00 \x00 명','생\x00 \x00 년\x00 \x00 월\x00 \x00 일','현\x00 \x00 \x00 \x00 주\x00 \x00 \x00 \x00 소','연\x00 \x00 \x00 \x00 락\x00 \x00 \x00 \x00 처']
i_d_Index2 =[]

for i in splitdata2:
    for j in infodata2:
        if i == j:
            i_d_Index2.append(splitdata2.index(i)+5)

for i in range(4):
    print(splitdata2[i_d_Index2[i]])
```

```python
import xlsxwriter
workbook = xlsxwriter.Workbook('test_pdf.xlsx')
worksheet = workbook.add_worksheet()

row = 0
col = 0
for i in i_d_Index2:
    worksheet.write(row,col,splitdata2[i-5])
    worksheet.write(row+1,col,splitdata2[i])
    col +=1
workbook.close()    
```
