```python
import xlsxwriter

workbook = xlsxwriter.Workbook('salary.xlsx')

worksheet = workbook.add_worksheet('sal_2022')

m_lst = ['C','D','E','F','G','H','I','J','K','L','M','N']
worksheet.write('A1','구분')
worksheet.write('B1','항목')
row = 2
for i in range(1,13):
    worksheet.write(0,row,str(i)+'월')                 #월 (C1:N1) 
    if i<=3:
        worksheet.write(1,row,1859000)            #봉급
        worksheet.write(6,row,93810)              #시간외수당(정액)
        worksheet.write(7,row,375240)             #시간외수당(실적)
    else:
        worksheet.write(1,row,1920900)
        worksheet.write(6,row,97360)
        worksheet.write(7,row,389440)          
    worksheet.write(2,row,120000)                 #특수지근무수당
    worksheet.write(3,row,155000)                 #직급보조비
    if (i<8 and i % 2 == 0) or (i>8 and i%2 ==1):
        worksheet.write(4,row,135270)             #영외급식비
    else:
        worksheet.write(4,row,139770)
    if i==1 or i == 9:
        worksheet.write(5,row,'= '+m_lst[i-1]+'2 * 0.6')      #명절휴가비
    worksheet.write(8,row,'=SUM('+m_lst[i-1]+'2:'+m_lst[i-1]+'8)')  #급여소계
    row += 1

worksheet.write('A2','급여')
worksheet.write('B2','봉급')
worksheet.write('B3','특수지근무수당')
worksheet.write('B4','직급보조비')
worksheet.write('B5','영외급식비')
worksheet.write('B6','명절휴가비')
worksheet.write('B7','시간외수당(정액)')
worksheet.write('B8','시간외수당(실적)')
worksheet.write('B9','급여소계')

workbook.close()
```
