##4.1 image_offset
- 오프셋 값은 픽셀단위로, 기본셀의 너비 또는 높이보다 클 수 있음
- 동일한 셀을 기준으로 둘 이상의 이미지를 정렬하려는 경우 유용

```python
import xlsxwriter


workbook = xlsxwriter.Workbook('tutorial_4_2.xlsx')
worksheet = workbook.add_worksheet()

worksheet.insert_image('B2', r'./python.jpg',{
    'x_offset' : 15, 'y_offset' : 10
})  # 이미지 위치 변경
workbook.close()
```

##4.3. image_scale
- x_scale 및 y_scale 파라미터는 수평 및 수직으로 이미지 크기를 조절하는데 사용 가능
  - worksheet.insert_image('B3', 'python.png', {'x_scale': 0.5, 'y_scale': 0.5})
- url 매개 변수는 하이퍼링크/URL을 이미지에 추가하는데 사용 가능
  - worksheet.insert_image('B4','python.png',{'url': 'https://python.org'})

```python
import xlsxwriter

workbook = xlsxwriter.Workbook('tutorial_4_3.xlsx')
worksheet = workbook.add_worksheet()

worksheet.insert_image('B2', r'./python.jpg',{
    'x_scale': 0.1, 'y_scale': 0.1, # 이미지 크기 변경
    'x_offset' : 15, 'y_offset' : 10,  # 이미지 위치 변경
    'url' : 'https://python.org'  # 링크 연결하기
})
workbook.close()
```

##4.4 web_image_data
- io.BytesIO형식으로 메모리 내 바이트스트림을 추가하는데 사용
  - worksheet.insert_image('B5','python.jpg', {'image_data': image_data})
- 일반적으로 URL에서 이미지를 삽입하는데 사용
  - url = 'https://python.org/logo.png' image_data = io.BytesIO(urlopen(url),read())
  - worksheet.insert_image('B5', url,{'image_data': image_data})
- 이미지 파일 이름이 필요하기 때문에 앞의 예저에서 파일이름은 url문자열에 추출

```python
import io

try:
    from urllib.request import urlopen
except:
    from urllib2 import urlopen


workbook = xlsxwriter.Workbook('tutorial_4_4.xlsx')
worksheet = workbook.add_worksheet()

#   "https://www.python.org/static/img/python-logo.png"
url = 'https://www.python.org/static/img/python-logo.png'

img = io.BytesIO(urlopen(url).read())

worksheet.insert_image('B5', url,{
    'image_data': img,
    'x_scale': 1.5, 'y_scale': 1.5
})

workbook.close()
```

##4.5 image_positioning<br>
Positioning 매개변수는 이미지를 배치하는 속성 값
- 셀과 함께 이동하고 크기를 조절 (positioning:1)
- (기본값) 셀과 함께 이동하지만 크기는 변동 없음 (positioning:2)
- 셀과 함께 움직이거나 크기를 저잘하지 않음 (positioning:3)

```python

workbook = xlsxwriter.Workbook('tutorial_4_5.xlsx')
worksheet = workbook.add_worksheet()

worksheet.insert_image('B5', r'./python.jpg',{
    'positioning':1   # 걸치고 있는 셀의 크기 줄일 시 같이 감소, 셀 잘라내기 후 이동 시 같이 이동
})
worksheet.insert_image('B25', r'./python.jpg',{
    'positioning':2   # 주변 셀 변경 시 변경 O, 걸치고 있는 셀은 영향 X
})
worksheet.insert_image('B45', r'./python.jpg',{
    'positioning':3   # 주변, 걸치고 있는 셀 변경과 관련 X, 이미지 혼자 떠있는 느낌
})

workbook.close()
```
