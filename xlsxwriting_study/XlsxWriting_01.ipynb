```python
import xlsxwriter

#엑셀 파일 생성
workbook = xlsxwriter.Workbook('tutorial_1_1.xlsx')

#파일 안 워크시트 생성
worksheet = workbook.add_worksheet('codingcamp')

#셀 안에 문자값 입력
worksheet.write('A1', 'A')
worksheet.write('B1', 'B')
worksheet.write('C1', 'C')
worksheet.write('D1', 'D')
worksheet.write('E1', 'E')

#셀 안에 문자값 입력
worksheet.write('A2', 1)
worksheet.write('B2', 2)
worksheet.write('C2', 3)
worksheet.write('D2', 4)
worksheet.write('E2', 5)

#좌표로 셀 안에 문자값 입력
worksheet.write(2, 0, 'a')
worksheet.write(2, 1, 'b')
worksheet.write(2, 2, 'c')
worksheet.write(2, 3, 'd')
worksheet.write(2, 4, 'e')

#좌표로 셀 안에 문자값 입력   
# worksheet.write(행=row값, 열=column값, 입력값)
worksheet.write(3, 0, 10)
worksheet.write(3, 1, 20)
worksheet.write(3, 2, 30)
worksheet.write(3, 3, 40)
worksheet.write(3, 4, 50)     # E4에 50 입력

workbook.close()  # open(), close() 파일생성 작업과 동일하게
# ctrl+H >>> 단어 찾아서 원하는 단어로 변경 가능
```

# 1.2 산술 연산

```python
import xlsxwriter

#엑셀 파일 생성
workbook = xlsxwriter.Workbook('tutorial_1_2.xlsx')

#파일 안 워크시트 생성
worksheet = workbook.add_worksheet('codingcamp')
worksheet.write('A1', 10)
worksheet.write('B1', 5)

#사칙연산
worksheet.write('A2', '=A1+B1') # 엑셀 수식
worksheet.write('B2', '=A1-B1')
worksheet.write('C2', '=A1*B1')
worksheet.write('D2', '=A1/B1')

#사칙연산 함수
worksheet.write('A3', '=SUM(A1,B1)') # 엑셀 수식
worksheet.write('B3', '=IMSUB(A1,B1)')
worksheet.write('C3', '=PRODUCT(A1,B1)')
worksheet.write('D3', '=QUOTIENT(A1,B1)')

workbook.close()
```

#1.3 비교연산

```python
import xlsxwriter

#엑셀 파일 생성
workbook = xlsxwriter.Workbook('tutorial_1_3.xlsx')

#파일 안 워크시트 생성
worksheet = workbook.add_worksheet('codingcamp')

#셀 안에 문자값 입력하기
worksheet.write('A1', 10)
worksheet.write('B1', 5)

#사칙연산
worksheet.write('A2', '=A1=B1') # 엑셀 수식
worksheet.write('B2', '=A1>B1')
worksheet.write('C2', '=A1>=B1')
worksheet.write('D2', '=A1<B1')
worksheet.write('E2', '=A1<=B1')
worksheet.write('F2', '=A1<>B1')  # python '!=' == '<>'

workbook.close()
```

#1.4 기타 연산자

```python
import xlsxwriter

#엑셀 파일 생성
workbook = xlsxwriter.Workbook('tutorial_1_4.xlsx')

#파일 안 워크시트 생성
worksheet = workbook.add_worksheet('codingcamp')

#셀 안에 문자값 입력
worksheet.write('A1', 10)
worksheet.write('B1', 20)
worksheet.write('C1', 30)
worksheet.write('D1', 40)
worksheet.write('E1', 50)

worksheet.write('A2', '=A1&B1') # 문자열 이어붙임
worksheet.write('B2', '=SUM(A1:E1)')
worksheet.write('C2', '=SUM(A1,B1,C1,D1,E1)')
worksheet.write('D2', '=SUM(A1:C1,D1:E1)') # 범위 병합
worksheet.write('E2', '=A1^2')  # 제곱
worksheet.write('F2', '=A1%')  # 백분율
worksheet.write('F2', '=AVERAGE(A1:E1)')  # 평균

workbook.close()
```

#1.5 집계함수

```python
import xlsxwriter

#엑셀 파일 생성
workbook = xlsxwriter.Workbook('tutorial_1_5.xlsx')

#파일 안 워크시트 생성
worksheet = workbook.add_worksheet('codingcamp')

score_data = [
    ['이름', '국', '영', '수' ],
    ['영한', 90, 80, 90],
    ['성훈', 91, 70, 80],
    ['경록', 92, 60, 70],
    ['재현', 93, 50, 80]
]

row = 0
col = 0

for 이름, 국, 영, 수 in score_data:
    worksheet.write(row, col, 이름)
    worksheet.write(row, col+1, 국)
    worksheet.write(row, col+2, 영)
    worksheet.write(row, col+3, 수)
    if row == 0:
        worksheet.write(row, col+4, '점수합계')
        worksheet.write(row, col+5, '최댓값')
        worksheet.write(row, col+6, '최솟값')
        worksheet.write(row, col+7, '평균')
        worksheet.write(row, col+8, '순위')
    else:
        worksheet.write(row, col+4, f'{국+영+수}') # 점수합계
        worksheet.write(row, col+5, f'{max(국, 영, 수)}') # 최댓값
        worksheet.write(row, col+6, f'{min(국, 영, 수)}') # 최솟값
        worksheet.write(row, col+7, f'{sum([국, 영, 수])/3}') # 평균
        worksheet.write(row, col+8, '=rank(H'+str(row+1)+',$H$2:$H$5,0)')  # 순위, 0 -> 내림차순
        worksheet.write(row, col+9, '=AVERAGE(B'+str(row+1)+':D'+str(row+1)+')')
    row += 1

workbook.close()
```
