##3.1 Column Chart

```python
import xlsxwriter

workbook = xlsxwriter.Workbook('tutorial_3_1.xlsx')
worksheet = workbook.add_worksheet()

# https://xlsxwriter.readthedocs.io/working_with_charts.html
chart = workbook.add_chart({'type':'column'})
bold = workbook.add_format({'bold':True})

# 데이터 만들기
chart.set_x_axis({'name': 'Test number'}) 
chart.set_x_axis({
    'name': 'Student Score',    # 차트 제목
    'name_font': {'size': 14, 'bold':True}
})

data = [[70, 64, 30],
        [45, 33, 90],
        [25, 44, 30]]

worksheet.write(0, 1, '국', bold)
worksheet.write(0, 2, '영', bold)
worksheet.write(0, 3, '수', bold)

worksheet.write(1, 0, '철수', bold)
worksheet.write(2, 0, '영희', bold)
worksheet.write(3, 0, '바둑이', bold)

worksheet.write_column('B2', data[0])   # 국어 값만 입력
worksheet.write_column('C2', data[1])
worksheet.write_column('D2', data[2])

chart.add_series({                      # series 는 범례??!
    'categories': '=Sheet1!$A$2:$A$4',  # 철수, 영희, 바둑이 -> 기준축(category)
    'values': '=Sheet1!$B$2:$B$4',      # 국어 점수 values
    'name': '=Sheet1!$B$1'              # 국어 
})

chart.add_series({
    'values': '=Sheet1!$C$2:$C$4',      # 영어 점수
    'name': '=Sheet1!$C$1'              # 영어 
}) 

chart.add_series({
    'values': '=Sheet1!$D$2:$D$4',      # 수학 점수
    'name': '=Sheet1!$D$1'              # 수학 
}) 

worksheet.insert_chart('A7', chart)

workbook.close()
```

```python
import pandas as pd
import xlsxwriter

# 데이터프레임
data = {
    '이름':['철수','영희','바둑이'],
    '국':[70, 64, 30],
    '영':[45, 33, 90],
    '수':[25, 44, 30]
}

df = pd.DataFrame(data)
df = df.set_index('이름')


# df['국'].mean()  # 판다스를 이용해 계산값을 구하고 엑셀로 만들면 속도가 훨씬 빠르다
writer = pd.ExcelWriter('tutorial_3_3_2_pandas_data.xlsx', engine = 'xlsxwriter')

# dir(writer) object확인 >>> 'book': workbook, 'sheets': worksheet와 같음

df.to_excel(writer, sheet_name='First')

workbook = writer.book
worksheet = writer.sheets['First']

chart = workbook.add_chart({'type':'line'})

chart.set_x_axis({
    'name': 'Student Score',    # 차트 제목
    'name_font': {'size': 14, 'bold':True}
})

cell_format = workbook.add_format()
cell_format.set_align('center')
cell_format.set_bold()

worksheet.write(4, 0, '평균', cell_format)
worksheet.write(4, 1, '=round(average(B2:B4),2)')
worksheet.write(4, 2, '=round(average(C2:C4),2)')
worksheet.write(4, 3, '=round(average(D2:D4),2)')

chart.add_series({ 
    'categories': '=First!$A$2:$A$5',  # 철수, 영희, 바둑이
    'values': '=First!$B$2:$B$5',      # 국어 점수
    'name': '=First!$B$1'              # 국어 
})

chart.add_series({
    'values': '=First!$C$2:$C$5',      # 영어 점수
    'name': '=First!$C$1'              # 영어 
}) 

chart.add_series({
    'values': '=First!$D$2:$D$5',      # 수학 점수
    'name': '=First!$D$1'              # 수학 
})

worksheet.insert_chart('A7', chart)

workbook.close()
writer.save()
```

##3.5 Pie Chart

```python
df = pd.read_excel('tutorial_3_4_1_pandas_data.xlsx')
df = df.set_index('이름')
df = df['철수':'바둑이']  # 평균값 행 제거

writer = pd.ExcelWriter('tutorial_3_5_1.xlsx', engine = 'xlsxwriter')

df.to_excel(writer, sheet_name = 'sheet')

workbook = writer.book
worksheet = writer.sheets['sheet']

chart = workbook.add_chart({'type':'pie'})
# pie차트는 한 변수(국어)만 출력이 가능
# 만약 pie차트 이용해 여러변수(국어, 영어, 수학) 출력하려면 
# 아래 doughnut_chart 사용

chart.add_series({
    'categories': '=sheet!$A$2:$A$4',  # 변수는 하나만 되기때문에 series 하나 입력
    'values': '=sheet!$B$2:$B$5',      
    'name': '=sheet!$B$1'              
})


worksheet.insert_chart('A7', chart)

workbook.close()
writer.save()
```
